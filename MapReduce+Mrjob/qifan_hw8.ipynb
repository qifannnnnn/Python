```python
import matplotlib.pyplot as pyt
darwin=open("darwin_word_counts.txt")
import re
find=re.compile(r"[\d]+")
count_list=list()
for i in darwin:
    count_list.append(int(re.findall(find,i)[-1]))
```

```python
# 横纵坐标顺序？？？？
import math
count_list.sort(reverse=True)
freq=count_list
rank=list(range(1,len(freq)+1))
freq_unique=[]
rank_unique=[]
for i in range(len(freq)):
    if not freq[i] in freq_unique:
        freq_unique.append(freq[i])
        rank_unique.append(rank[i])
pyt.plot(list(map(math.log,rank_unique)),list(map(math.log,freq_unique)))
pyt.xlabel("Log Word Rank")
pyt.ylabel("Log Word Frequency")
pyt.title("Zip's Law")
pyt.savefig("zipf.pdf")
__=pyt.show()
```

#### The plot is indeed approximately a straight line.
##### (I noticed that there are some words with the same frequency and to handle this, I gave the words with the same frequency the same rank as well. The plot looks more linear this way.
##### A simple example: if the list contains "and"  3,  "the" 3 and "time" 2.  "and" and "the" both rank first while "time" ranks third. )

```python
import re
import scipy.stats as stats
import math
regex=re.compile("[-]{0,1}[\d]+\.{0,1}[\d]*")
summary=open("summary_large.txt")
labels,means,numbers,variances,yerr=[],[],[],[],[]
for i in summary:
    temp=re.findall(regex,i)
    labels.append(temp[0])
    numbers.append(int(temp[1]))
    means.append(float(temp[2]))
    variances.append(float(temp[3]))
    yerr.append(math.sqrt(variances[-1])/math.sqrt(numbers[-1])*1.96)
labels_order=labels[0:2]+labels[4:]+labels[2:4]
means_order=means[0:2]+means[4:]+means[2:4]
yerr_order=yerr[0:2]+yerr[4:]+yerr[2:4]
pyt.errorbar(labels_order,means_order,yerr=yerr_order,fmt="o")
pyt.xlabel("labels")
pyt.ylabel("confidence interval range")
pyt.title("Confidence Intervals")
pyt.savefig("populations.pdf")
__=pyt.show()
```
